## Exercise 1
### Write a Python program to get the difference between a given number and 17, if the number is greater than 17 return double the absolute difference.


```python
x = float(input("Input Number:"))

if x > 17:
    print(x*2)
else:
    print(17-x)
```

## Exercise 2
### Write a Python program to calculate the sum of three given numbers, if the values are equal then return thrice of their sum.


```python
x = float(input("Input 1st Number:"))
y = float(input("Input 2nd Number:"))
z = float(input("Input 3rd Number:"))

if x==y==z:
    print((x+y+z)*3)
else:
    print(x+y+z)
```

## Exercise 3
### Write a Python program which will return true if the two given integer values are equal or their sum or difference is 5.


```python
x = float(input("Input 1st Number:"))
y = float(input("Input 2nd Number:"))

if x==y:
    print(True)
elif x+y==5:
    print(True)
elif x-y==5:
    print(True)
else:
    print(False)
```

## Exercise 4
### Write a Python program to sort three integers without using conditional statements and loops.


```python
a = float(input("Input 1st Number:"))
b = float(input("Input 2nd Number:"))
c = float(input("Input 3rd Number:"))

x= min(a,b,c)
z= max(a,b,c)
y=(((a+b+c)-z)-x)
print("Min value:",x)
print("Middle value:",y)
print("Max value:",z)
```

## Exercise 5
### Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.


```python
x = int(input("Enter a number:"))
y = list(range(1,x))
k = 0;
for i in y:
    z = i**3
    k = k + z
print(k)
```
